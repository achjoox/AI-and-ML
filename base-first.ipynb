```python
import keras
import keras_nlp
```

```python
model = keras_nlp.models.GemmaCausalLM.from_preset("gemma_1.1_instruct_2b_en")
```

```python
model.summary()
```

```python
prompt_template="""
You are a head of marketing division. Make me an email to promote a product with called Jackys.
Target of a marketing is girl who like shopping
"""
```

```python
model.generate(prompt_template, max_length=2048)
```

```python
START_USER = '<start_of_turn>user\n'
START_MODEL = 'start_of_turn>model\n'
END_TURN = '<end_of_turn>\n'
model.generate(START_USER+prompt_template+END_TURN+START_MODEL, max_length=2048)
```
